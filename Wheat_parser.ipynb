## Wheat Data Analysis

```python
import pandas as pd
```

```python
df = pd.read_excel('Wheat_Prices.xlsx')
df
```

```python
df.columns
```

```python
#getting the data into tidy data shape
df2 = df.set_index(['Date']).stack().reset_index().rename(columns={'level_1': 'Origin', 0: 'Price'})
df2
```

```python
legend = pd.read_excel('WHEAT_Legend.xlsx')
legend
```

```python

df2['Commodity'] = 'Wheat'
df2
#getting rid of the time from the date
df2['Date'] = pd.to_datetime(df2['Date']).dt.date
```

```python
#merging the legend data into the df2
df2 = df2.merge(legend[['Origin', 'Location', 'Country', 'Continent', 'FOB_Price_Type', 'Currency/Unit']], on='Origin', how='left')
df2
```

```python
df2.to_csv('Wheat_Prices_CSV.csv', index=False)

df2.to_json('Wheat_Prices_JSON.json', orient='records')
```
